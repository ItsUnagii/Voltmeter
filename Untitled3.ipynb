```python
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Define the starting time (03-11 12:43 PM)
start_time = datetime.strptime('03-11 12:43', '%m-%d %H:%M')

try:
    with open('AudioMothR_3-11_12-43.txt', 'r') as file:
        x_values = []
        y_values = []
        for line in file:
            line = line.strip()
            if not line:  # skip empty lines
                continue
            try:
                hours_since_start, y = line.split(',')
                hours_since_start = float(hours_since_start.strip())  # convert to float for hours
                y_values.append(float(y.strip()))  # voltage to float
                # calculate actual datetime by adding hours to the start_time
                actual_time = start_time + timedelta(hours=hours_since_start)
                x_values.append(actual_time)
            except ValueError:
                print(f"Skipping invalid line: {line}")

except FileNotFoundError:
    print("Error: AudioMothR_3-11_12-43.txt not found.")
except ValueError:
    print("Error: Invalid data format in AudioMothR_3-11_12-43.txt.")
else:
    plt.figure(figsize=(10, 6))
    plt.plot(x_values, y_values)
    plt.xlabel('Date and Time')
    plt.ylabel('Voltage')

    plt.xticks(rotation=45)

    plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%m-%d %H:%M'))

    plt.axvline(x=(datetime.strptime('03-13 00:30', '%m-%d %H:%M')), color='red', linestyle='--', label="Start time")
    plt.tight_layout()
    plt.show()
```
